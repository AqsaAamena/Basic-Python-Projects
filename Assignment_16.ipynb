Q7. Create a heatmap to show the correlation between numerical variables.
Use sns.heatmap()
Annotate the values and use a suitable color palette

```python
import matplotlib.pyplot as plt
```

```python
# Select only the numerical columns for correlation
numerical_df = df[['age', 'bmi', 'children', 'expenses']]
# Calculate the correlation matrix
corr = numerical_df.corr()
# Create the heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True,cmap='rocket')
# Add title
plt.title('Correlation Heatmap of Numerical Variables')
plt.show()
```

Q8. Perform the following analysis on the charges column:
Find the average insurance charges

Find the maximum and minimum charges

Group by smoker and compare the average charges for smokers vs non-
smokers

```python
# Calculate statistics
avg = df['expenses'].mean()
max_val = df['expenses'].max()
min_val = df['expenses'].min()
print(f"The average insurance charges is: ${avg.round(2)}")
print(f"The maximum insurance charges is: ${max_val}")
print(f"The minimum insurance charges is: ${min_val}")
print("-" * 50)
# Group by smoker and compare average charges
print("Average charges for smokers vs non-smokers:")
grp = df.groupby('smoker')['expenses'].mean().round(2)
print(grp)
```

Q9. Create a boxplot to show the distribution of charges with respect to smoker
