Q7. (Matrices and Operations)

Q9. (Combined - Random + Reshape + Statistics)
Generate a 1D array of 20 random integers between 1 and 50 using
np.random.randint().
Reshape it into a 4x5 matrix.
Print the sum, mean, and standard deviation of the matrix.
Find the maximum value in each row.

```python
import numpy as np
arr=np.random.randint(1,51,20)
print("1D array:")
print(arr)
reshape=arr.reshape(4,5)
print("Reshaped 4x5 matrix:")
print(reshape)
sum=reshape.sum()
print("sum of matrix:-",sum)
mean=reshape.mean()
print("mean of matrix:-",mean)
std=np.std(reshape)
print("Standard deviation:-",std)
max=reshape.max(axis=1)
print("maximum value in each row:-",max)
```

Q10. (Mini Project - NumPy Application)
Create a Simple Statistics Calculator using NumPy:
Write a program that:
Asks the user how many numbers they want to generate.
Generates that many random numbers between 10 and 100 using
np.random.randint().
Stores them in a 1D array.
Prints:
The array
Mean, Median, Standard Deviation
Minimum and Maximum
Reshapes into a 2D array (if possible) and prints row-wise sum.
Use proper comments and exception handling where needed.

```python
import numpy as np
num=int(input("How many numbers you want to generate:"))
#use to generate random integers
arr=np.random.randint(10,101,num)
print("Array:",arr)
# methods to print basic statistics
mean=arr.mean()
print("Mean:",mean)
median=np.median(arr)
print("Median:",median)
std=np.std(arr)
print("Standard Deviation:",std)
min=arr.min()
print("Minimum:",min)
max=arr.max()
print("Maximum:",max)
factors = [i for i in range(2, num) if num % i == 0]
try:
  if len(factors) > 0:
    rows = factors[len(factors) // 2]
    cols = num // rows
    two_d = arr.reshape(rows, cols)
    print(f"\nArray reshaped into a {rows}x{cols} matrix:\n", two_d)
    row_sum = two_d.sum(axis=1)
    print("Row-wise sum:", row_sum)
  else:
    raise ValueError
except ValueError:
  print(f"ValueError: cannot reshape array of size {num} into shape ({rows},{cols})")
```
