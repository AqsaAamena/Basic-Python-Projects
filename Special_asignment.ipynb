```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("meesho ForwardReports.csv")

# Display first few records
df.head(10)
```

```python
space="="*50
#Check dataset shape.
print("Shape of Dataset:-",df.shape)
print(space)

#Display column names.
print("Column Names:-\n",df.columns)
print(space)

#Check data types.
print("Data Types:-")
print(df.dtypes)
print(space)

#Identify missing value
print("Missing Values:-")
print(df.isnull().sum())
```

```python
#Generate summary statistics.
df.describe().round(2)
```

```python
# Visualization Style
sns.set(style="whitegrid", palette="muted")
plt.figure(figsize=(5,4))

# 1.Bar Chart – Order Status Distribution
plt.figure(figsize=(7,5))
sns.countplot(x='order_status', data=df, order=df['order_status'].value_counts().index)
plt.title("Order Status Distribution", fontsize=14, fontweight='bold')
plt.xlabel("Order Status")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

```

```python
print()
# 2.Histogram – Meesho Price Distribution
plt.figure(figsize=(7,5))
sns.histplot(df['meesho_price'], bins=30, kde=True, color='purple')
plt.title("Distribution of Meesho Price", fontsize=14, fontweight='bold')
plt.xlabel("Meesho Price (₹)")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3.Scatter Plot – GST vs Price
plt.figure(figsize=(7,5))
sns.scatterplot(x='price', y='gst_amount', hue='order_status', data=df)
plt.title("GST vs Price by Order Status", fontsize=14, fontweight='bold')
plt.xlabel("Price (₹)")
plt.ylabel("GST Amount (₹)")
plt.show()
```

```python
# 4.Box Plot – Price by State
plt.figure(figsize=(10,6))
sns.boxplot(x='state', y='price', data=df)
plt.title("Price Distribution by State", fontsize=14, fontweight='bold')
plt.xlabel("State")
plt.ylabel("Price (₹)")
plt.xticks(rotation=90)
plt.show()
plt.tight_layout()
```

```python
# 5.Heatmap – Correlation Matrix
plt.figure(figsize=(7,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm_r')
plt.title("Correlation Heatmap", fontsize=14, fontweight='bold')
plt.show()
```

```python
# 6.Pie Chart – Shipping Charges Breakdown
plt.figure(figsize=(5,5))
shipping_counts = (df['shipping_charges_total'] > 0).value_counts()
plt.pie(shipping_counts, labels=['With Charges', 'Free Shipping'], autopct='%1.1f%%')
plt.title("Shipping Charges Breakdown", fontsize=14, fontweight='bold')
plt.show()
```

```python
# Key Insights Summary
print("\n============== Key Insights===========")
print("- Delivered orders dominate, but Returns/RTO are significant.")
print("- Most prices cluster between ₹500–₹2000.")
print("- GST correlates strongly with product price.")
print("- Free shipping is common (~70% of orders).")
print("- States like Uttar Pradesh, West Bengal, and Tamil Nadu have high order volumes.")
```
